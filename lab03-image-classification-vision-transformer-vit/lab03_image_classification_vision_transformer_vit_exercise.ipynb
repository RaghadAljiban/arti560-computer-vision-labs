```python
# Import Libraries
import os
import keras
import tensorflow as tf
import tensorflow_text as tf_text
import keras_hub
import requests
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load ViTImageClassifierPreprocessor (vit_base_patch16_224_imagenet preset)
preprocessor = keras_hub.models.ViTImageClassifierPreprocessor.from_preset(
    "vit_base_patch16_224_imagenet"
)
```

```python
# Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)
model = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet",
    load_weights=True,
)
```

```python
from PIL import Image

image_paths = [
    "/content/computer_mouse.jpg",
    "/content/dough.jpg",
    "/content/espresso.jpg",
    "/content/golden_retriever.jpg",
    "/content/sport_car.jpg"
]

images = [Image.open(path) for path in image_paths]

print("Images loaded successfully")
```

```python
# Predict classes
url = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"

response = requests.get(url)
class_names = response.text.strip().split("\n")

print("Number of classes:", len(class_names))
print("First 10 classes:", class_names[:10])

results = []

for path in image_paths:
    img = Image.open(path).convert("RGB")
    x = np.asarray(img, dtype=np.uint8)[None, ...]

    inp = preprocessor(x)
    logits = model(inp, training=False).numpy()

    pred_id = int(np.argmax(logits, axis=-1)[0])
    pred_label = class_names[pred_id]

    results.append({
        "Image File": os.path.basename(path),
        "Predicted Label": pred_label,
    })

df = pd.DataFrame(results)
df
```
